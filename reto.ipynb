```python
df = pd.read_csv('covid19_tweets.csv')

#formato de la fecha
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S')

df['fecha'] = df['date'].dt.date


tweets_por_fecha = df.groupby('fecha').size()

# gr√°fico
plt.figure(figsize=(6, 3))
plt.plot(tweets_por_fecha.index, tweets_por_fecha.values, marker='o', linestyle='-')
plt.title('N√∫mero de Tweets por D√≠a')
plt.xlabel('Fecha')
plt.ylabel('N√∫mero de Tweets')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# dfi es el dataframe en donde se almacenan los usuarios que detectamos como no bots, aunque hay cierta incertidumbre de que a√∫n existan algunos que pudieron pasar los filtros.
dfi=df.copy()

#dfb es el dataframe donde se almacenan los usuarios que detectamos como bots, debido a nuestros par√°metros utilizados.
dfb=df.copy()

#dfn es el dataframe donde se almacenan los usuarios que podemos asegurar respecto a nuestros parametros que no son bots
dfn=df.copy()
```

```python
#Dataframe de usuarios que sabemos que NO SON BOTS respecto a nuestros par√°metros, pero con cierta incirtedumbre
dfi.drop(dfi[(dfi['user_verified']==False) & (dfi['user_followers'] > dfi['user_friends']*4) & (dfi['Year']>2015)].index, inplace=True)
dfi
```

```python
#Dataframe de usuarios que sabemos al 100% que NO SON BOTS respecto a nuestros par√°metros
dfn.drop(dfn[(dfn['user_verified']==False) & (dfn['user_followers'] > dfn['user_friends']*4) & (dfn['Year']>2015)].index, inplace=True)
dfn.drop(dfn[(dfn['user_verified'] == False)].index, inplace=True)
dfn.drop(dfn[(dfn['user_followers'] > dfn['user_friends']*4)].index, inplace=True)
dfn.drop(dfn[(dfn['Year']>2015)].index, inplace=True)
dfn
```

```python
# Dataframe en donde se almacenan los usuarios considerados como bots

dfb.drop(dfb[(dfb['user_verified'] == True)].index, inplace=True)
dfb.drop(dfb[(dfb['user_followers'] < dfb['user_friends']*4)].index, inplace=True)
dfb.drop(dfb[(dfb['Year']<2015)].index, inplace=True)
dfb
```
